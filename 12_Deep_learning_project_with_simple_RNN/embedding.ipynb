```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

- one hot encoding happens with hash mapping that's why we need to give vocabulary size explicitily.
- where as tokennizer is different scenario, it takes the values automatically depending on the size of unique words.
- we should not give vocabulary size less than actual unique size in one hot because, you might end up having same hash number for 2 different words. which will be a issue again.

```python
voc_size = 10000

one_hot_sent = [one_hot(words,voc_size) for words in sent]
one_hot_sent
```

```python
sent_length = 8
padded_sent = sequence.pad_sequences(one_hot_sent, padding='pre', maxlen=sent_length)
padded_sent
```

- dim/output dim is the size of the vector for each word.
- you must give dim, because input_length only tells how long the sequence is â€” not how you want to represent the words.

```python
dim=10

model = Sequential()
#Embedding(input_dim=voc_size, output_dim=10, input_length=sent_length)
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(padded_sent)
```

```python
model.predict(padded_sent[0])
```

- as we can see now we are getting each words vectors as 10 dimension values for 8 words.
- as we have zero for first 3 words, we are getting same vectors values for them.
