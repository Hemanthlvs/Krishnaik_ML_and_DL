#### The above diagram looks bit right skewed as you can see more records at the age 20-30. so we can use median instead of mode that's a best choice.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# Q-Q Plot
stats.probplot(df['age'].dropna(), dist="norm", plot=plt)
plt.title("Q-Q Plot for 'age'")
plt.show()
```

#### It's better to use both histogram and Q-Q plot to get better understanding.

```python
df['Age_mean']= df['age'].fillna(df['age'].mean())
df
```

# When & Why to Check Histogram and Q-Q Plot

## 1. Before Filling Missing Values
- Check if data is normal (bell-shaped) or skewed.
- Normal → fill missing with **mean**.
- Skewed → fill missing with **median**.

## 2. Before Using Machine Learning Models
- Some models assume normal data (e.g., linear regression).
- If not normal, consider transforming data (like log).

## 3. For Feature Engineering & Understanding Data
- Histogram shows shape, skewness, outliers.
- If skewed, **transform data (applying a math function like log, square root, or Box-Cox to “pull in” the long tail and make the data more balanced (more normal))** or **create new features(create new features from the original data to better represent it — like categorizing age groups instead of using raw age)**.
- If multiple peaks, data may have different groups.
	- When the histogram shows more than one peak (multi-modal), it often means the data comes from different groups or populations mixed together.
	- For example, a dataset of heights might have two peaks if it includes both adults and children.

## 4. After Transformations
- Check if data is closer to normal after changes.

## 5. Checking Model Errors
- Residuals should be normal for some models.

### 2. Median value Imputation
### This is used when we have a outliers in the dataset

```python
df['age_median'] = df['age'].fillna(df['age'].median())
df
```

```python
df[['age', 'age_median']]
```

### 3. Mode Imputation
### This is mostly used for categorical values

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
df['mode_embarked'] = df[df['embarked'].notna()]['embarked'].mode()[0]
df
```

```python
df[['mode_embarked', 'embarked']]
```

```python
df[df['embarked'].isnull()]
```

```python
# Alternative way

df['embarked_mode'] = df['embarked'].fillna(df['embarked'].mode()[0])
df
```
