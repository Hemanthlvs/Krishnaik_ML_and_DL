```python
import pandas as pd
```

```python
data = pd.read_csv("smsspamcollection.csv", sep="\t", names=["labels","message"])
data
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corp = []
for i in range(0,len(data)):
    removed_special_characters = re.sub('[^a-zA-Z]',' ',data['message'][i])
    lower_case_data = removed_special_characters.lower()
    listed_data = lower_case_data.split()
    stemming_data = [ps.stem(word) for word in listed_data if not word in stopwords.words('english')]
    stemming_data = ' '.join(stemming_data)
    corp.append(stemming_data)
```

```python
corp
```

```python
# Bag Of Words

from sklearn.feature_extraction.text import CountVectorizer
# max_features = max number of features based on frequency
# binary = this will convert values to 0 and 1. even if the values are greater than 1 they convert to 1.
cv = CountVectorizer(max_features=500, binary=True)
X = cv.fit_transform(corp).toarray()
X
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
X
```

```python
# to check the vocabularies
cv.vocabulary_
```

```python
# N-grams

from sklearn.feature_extraction.text import CountVectorizer
# cv_1 = creates unigrams and bigrams
# cv_2 = creates unigrams, bigrams and trigrams
# cv_3 = creates bigrams and trigrams
# cv_4 = you can also create trigrams alone
# cv_5 = you can also create bigrams alone
cv_1 = CountVectorizer(max_features=500, binary=True, ngram_range=(1,2))
cv_2 = CountVectorizer(max_features=500, binary=True, ngram_range=(1,3))
cv_3 = CountVectorizer(max_features=500, binary=True, ngram_range=(2,3))
cv_4 = CountVectorizer(max_features=500, binary=True, ngram_range=(3,3))
cv_5 = CountVectorizer(max_features=500, binary=True, ngram_range=(2,2))
```
