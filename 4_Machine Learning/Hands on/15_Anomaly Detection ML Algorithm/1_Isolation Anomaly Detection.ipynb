```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
health_df = pd.read_csv('healthcare.csv')
health_df
```

```python
plt.scatter(health_df.iloc[:,0],health_df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest

IF = IsolationForest(contamination=0.2)
IF.fit(health_df)
predictions = IF.predict(health_df)
```

```python
predictions
```

```python
plt.scatter(health_df.iloc[:,0],health_df.iloc[:,1],c=predictions)
```

```python
X=health_df.values

index = np.where(predictions<0)
plt.scatter(health_df.iloc[:,0], health_df.iloc[:,1])
plt.scatter(X[index,0],X[index,1],edgecolors="r")
```
