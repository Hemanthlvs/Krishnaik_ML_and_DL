```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import make_classification

x,y = make_classification(n_samples=1000, n_features=2, n_classes=2, 
                          n_clusters_per_class=2, n_redundant=0)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0], y=pd.DataFrame(x)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state = 10)
```

```python
x_test
```

```python
from sklearn.svm import SVC

linear = SVC(kernel='linear')

linear.fit(x_train, y_train)
```

```python
y_pred_linear = linear.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, y_pred_linear))
print(confusion_matrix(y_test, y_pred_linear))
```

```python
poly = SVC(kernel='poly')
poly.fit(x_train, y_train)
```

```python
y_pred_poly = poly.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, y_pred_poly))
print(confusion_matrix(y_test, y_pred_poly))
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(x_train, y_train)
```

```python
y_pred_rbf = rbf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, y_pred_rbf))
print(confusion_matrix(y_test, y_pred_rbf))
```

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(x_train, y_train)
```

```python
y_pred_sigmoid = sigmoid.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, y_pred_sigmoid))
print(confusion_matrix(y_test, y_pred_sigmoid))
```

## Hyperparameter tuning with SVC

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf', 'poly', 'linear', 'sigmoid']}
```

```python
## verbose=3 will provide detailed logging information during the grid search process.

## refit=True will retrain the model with the best parameters on the entire dataset after the grid search is complete.

grid = GridSearchCV(SVC(), param_grid=param_grid, refit=True, cv=5, verbose=3)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred_grid = grid.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_grid))
print(confusion_matrix(y_test, y_pred_grid))
```
