```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris['DESCR'])
```

```python
x = pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width','petal length','petal width'])
```

```python
x
```

```python
y = iris['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=20)
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()

DTC.fit(x_train, y_train)
```

```python
# we can alsoe visualise the tree

from sklearn.tree import plot_tree
plt.figure(figsize=(15,10))
plot_tree(DTC, filled=True)
```

### By observing the diagram we can decide where to stop the tree by not going into the depth this is called post pruning. 
### since we are doing this after tree created.

```python
y_pred = DTC.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Now lets do post pruning

```python
params = {
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3]
}
```

```python
from sklearn.model_selection import GridSearchCV

decision_tree_grid = GridSearchCV(estimator=DTC, param_grid=params, cv=5, scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')

decision_tree_grid.fit(x_train,y_train)
```

```python
decision_tree_grid.best_params_
```

```python
decision_tree_grid.best_score_
```

```python
grid_pred = decision_tree_grid.predict(x_test)
```

```python

print(confusion_matrix(grid_pred, y_pred))
print(classification_report(grid_pred, y_pred))
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(grid_pred,y_pred)

score
```
